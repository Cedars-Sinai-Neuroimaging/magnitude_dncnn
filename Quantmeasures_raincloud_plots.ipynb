```python
%matplotlib inline
import pandas as pd
import ptitprince as pt
import matplotlib.pyplot as plt
```

```python
pt.RainCloud
```

```python
df = pd.read_excel('P_52_raincloud.xlsx')
print(df.head(5))
```

```python
f,ax = plt.subplots(figsize=(8,6))
ax = pt.RainCloud(x = "Processing", y = "PSNR", 
                  data = df, 
                  width_viol = .4,
                  width_box = .4, ax=ax, pointplot=True, point_size=5,
                  orient = 'v')
```
